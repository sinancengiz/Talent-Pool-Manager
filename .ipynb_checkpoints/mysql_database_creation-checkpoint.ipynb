```python
# import necessary libraries
import pandas as pd
from sqlalchemy import create_engine
import pymysql
pymysql.install_as_MySQLdb()
engine = create_engine('mysql://localhost:3306', echo=False)
```

```python
path = "Resources/data.xlsx"
employee_df = pd.read_excel(path)
employee_df.head()
```

```python
employee_df.to_sql('employees', con=engine, index=True, index_label="id")
```
